```python
#Checking feature with very high Correlation
#Drop highly correlated features
#Checking feature with lowest correction coefficient wrt. '% of Cap'
#Magnified Correlation Heatmap
#Variance Inflation Factor for 9 features with highest correlation wrt, '% of Cap' to check for Multicollinearity
#Check for Multicollinearity after removing features with high VIF
#Drop features with high Multicollinearity
```

```python
#Variance Inflation Factor for 9 features with highest correlation wrt, '% of Cap' to check for Multicollinearity
x = regular[['PTS', 'MP', 'FT', '2P', 'GS', 'TOV', 'AST', 'DRB', 'STL']]
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["features"] = x.columns

vif.round(1)
```

```python
#Check for Multicollinearity after removing features with high VIF
x = regular[['PTS', 'GS', 'AST', 'DRB', 'STL']]
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["features"] = x.columns

vif.round(1)
```

```python
#Drop features with high Multicollinearity
regular.drop(['FT', '2P', 'MP', 'TOV', '% of Cap'], axis = 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#Splitting data into Train and Test
X = basic
y = df['% of Cap']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

#Fit data with Regression model
linear_basic = LinearRegression()
linear_basic.fit(X_train, y_train)
y_hat = linear_basic.predict(X_test)

#Calculating r-squared
r2_score(y_test, y_hat)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.metrics import SCORERS
SCORERS.keys()
```

```python
X = new_df
y - df['% of Cap']

linear_model = LinearRegression()
cv = cross_validate(linear_model, X, y, cv = KFold(10, shuffle = True, random_state = 0), scoring = 'r2')
print('Mean Rsquared: ', np.mean(cv['test_score']))
print('STD Rsquared: ', np.std(cv['test_score']))
```

```python
poly = PolynomialFeatures(degree = 2, include_bias = False)
X_poly = poly.fit_transform(X)

cross_val_score(linear_model, X_poly, y, cv = KFold(10, shuffle = True), scoring = 'r2')
```

```python
ridge = Ridge()
param_grid = [{'alpha': range(1,50)}]

gridsearch = GridSearchCV(ridge, param_grid, cv = KFold(10, shuffle = True, random_state = 0), scoring = 'r2')
gridsearch.fit(X, y)
gridsearch.best_estimator_
```

```python
ridge = Ridge(alpha = 21)
scores = cross_validate(ridge, X, y, cv = KFold(10, shuffle = True, random_state = 0), scoring = 'r2')
print('Mean: ', np.mean(scores['test_score']))
print('STD: ', np.std(scores['test_score']))
```

```python
lasso = Lasso()
param_grid = [{'alpha': [0.0001, 0.001, 0.01, 0.1, 1]}]

gridsearch = GridSearchCV(lasso, param_grid, cv = KFold(10, shuffle = True, random_state = 0), scoring = 'r2')
gridsearch.fit(X, y)
gridsearch.best_estimator_
```

```python
gridsearch.best_estimator_.fit(X,y)
```

```python
lasso = Lasso(alpha = 0.001)
scores = cross_validate(lasso, X, y, cv = KFold(10, shuffle = True, random_state = 0), scoring = 'r2',
                         return_estimator = True, return_train_score = True)
print('Mean: ', np.mean(scores['test_score']))
print('STD: ', np.std(scores['test_score']))
```

```python
scores
```

```python
irrelevant_feats = []
for model in scores['estimator']:
    feat_coef = pd.Series(model.coef_, index = X.columns)
    irrelevant_feats += list(feat_coef[feat_coef.abs() == 0].index)
```

```python
irrelevant_feats
```

```python
values, counts = np.unique(irrelevant_feats, return_counts = True)
unused = pd.Series(counts, index = values)[counts >= 8]
```
